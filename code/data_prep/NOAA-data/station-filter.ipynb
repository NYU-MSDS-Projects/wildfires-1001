```python
import pandas as pd
import numpy as np
import re
```

### Read Stations into DATAFRAME

```python
stations_filename = "C:\\Users\\jades\\1001 Intro to Data Science Notebooks\\Project\\Data\\ghcnd-stations.txt"
stations_infile = open(stations_filename, 'r') 
station_lines = stations_infile.readlines() 
stations_infile.close()

df_list = []
for line in station_lines:
    char_split_list = []
    char_split_list.append(line[:11])
    char_split_list.append(line[12:20])
    char_split_list.append(line[21:30])
    char_split_list.append(line[31:37])
    char_split_list.append(line[38:40])
    char_split_list.append(line[41:71])
    char_split_list.append(line[72:75])
    char_split_list.append(line[76:79])
    char_split_list.append(line[80:85])
    char_split_list = [x.strip() for x in char_split_list]
    df_list.append(char_split_list)
    
```

```python
stations_df = pd.DataFrame(df_list, columns = ["ID","LATITUDE","LONGITUDE","ELEVATION","STATE","NAME","GSN FLAG","HCN/CRN FLAG","WMO ID"])
CA_stations_df= stations_df.where(stations_df["STATE"]=="CA").dropna()
CA_stations_df
```

```python
CA_stations_ID_list = CA_stations_df["ID"].unique()
```

### Read Inventory into DATAFRAME

```python
inventory_filename = "C:\\Users\\jades\\1001 Intro to Data Science Notebooks\\Project\\Data\\ghcnd-inventory.txt"
inventory_infile = open(inventory_filename, 'r') 
inventory_lines = inventory_infile.readlines() 
inventory_infile.close()

df_list = []
for line in inventory_lines:
    char_split_list = []
    char_split_list.append(str(line[:11].strip()))
    char_split_list.append(float(line[12:20]))
    char_split_list.append(float(line[21:30]))
    char_split_list.append(str(line[31:35].strip()))
    char_split_list.append(int(line[36:40]))
    char_split_list.append(int(line[41:45]))
    #char_split_list = [x.strip() for x in char_split_list]
    df_list.append(char_split_list)
    
```

```python
inventory_df = pd.DataFrame(df_list, columns = ["ID","LATITUDE","LONGITUDE","ELEMENT","FIRSTYEAR","LASTYEAR"])
CAin_df = inventory_df.where(inventory_df["ID"].isin(CA_stations_ID_list)).dropna()
CAin_df
```

```python
CAin_df["ELEMENT"].value_counts()[:10]
```

```python
time_CAin_df = CAin_df.where((CAin_df["LASTYEAR"] == 2020) & (CAin_df["FIRSTYEAR"] <= 1990)).dropna()
time_CAin_df
```

```python
combined_CA_element_list = time_CAin_df[["ID","ELEMENT"]].groupby('ID').agg(lambda x: x.tolist())#.where(time_CAin_df["ELEMENT"].isin(element_list)).dropna()
combined_CA_element_list = combined_CA_element_list.reset_index()
combined_CA_element_list
```

```python
element_list = ["TMAX","TMIN","PRCP"]

subset_mask = np.zeros(len(combined_CA_element_list["ELEMENT"]))

for i,j in enumerate(combined_CA_element_list["ELEMENT"]):
     subset_mask[i] = set(element_list).issubset(set(j))

combined_CA_element_list["ELEMENT MASK"] = subset_mask
combined_CA_element_filter = combined_CA_element_list.where(combined_CA_element_list["ELEMENT MASK"] ==1).dropna()
combined_CA_element_filter
```
