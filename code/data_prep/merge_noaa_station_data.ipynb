# Merge NOAA Station Data

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
fileNames = pd.read_csv('C:\\Users\\jades\\1001 Intro to Data Science Notebooks\\Project\\Data\\Station List v2.csv')
```

```python
#Set the directory location
gitdir = 'C:\\Users\jades\\1001 Intro to Data Science Notebooks\\Project\\wildfires-1001\\'
subdir = 'data\\raw_data\\NOAA-station-data\\'
```

```python
#Create a master table with all stations
ind = 1

df_master = pd.read_csv(gitdir + subdir + fileNames['ID'][0] + '.csv')

for fileName in fileNames['ID'][1:]:
    df = pd.read_csv(gitdir + subdir + fileName + '.csv')
    
    ind += 1
    
    print('Merging ' + fileName + ' (Index: ' + str(ind) + ')')
    df_master = pd.concat((df_master, df))
    print('New master df size: ' + str(len(df_master)))
```

```python
#Filter the data

#Cap the years we are looking at
df_master_trim = df_master[df_master['YEAR'] >= 1990]

#Drop the sparsely populated columns
df_master_trim = df_master_trim.drop(['DAPR', 'MDPR', 'SNWD', 'WT01', 'WT03', 'WT04', 'WT05', 'WT07', 'WT11',
       'DASF', 'MDSF', 'WT06', 'WT08', 'WT14', 'WT16', 'WT18', 'DAEV', 'EVAP',
       'MDEV', 'WDMV', 'WT09', 'DAWM', 'MDWM', 'MNPN', 'MXPN', 'ACMH', 'ACSH',
       'FRGT', 'THIC', 'WDF1', 'WSF1', 'WSFG', 'WT02', 'SN02', 'SN03', 'SN32',
       'SN33', 'SN35', 'SX02', 'SX03', 'SX32', 'SX33', 'SX35', 'WT10', 'AWND',
       'FMTM', 'PGTM', 'WDFG', 'WDFM', 'WESD', 'WSFM', 'TAVG', 'AWDR', 'WSFI',
       'TSUN', 'WDF2', 'WDF5', 'WSF2', 'WSF5', 'WT13', 'WV03', 'WV07', 'WV20',
       'WT17', 'WT19', 'WT21', 'WT22', 'WV01', 'WT12', 'WT15', 'WDFI', 'PSUN',
       'FRGB', 'FRTH', 'GAHT', 'WESF'], axis=1)

#Replace the dummy nulls with actual nulls
df_master_trim.replace(-9999, np.nan, inplace=True)
```

```python
gitdir = 'C:/Users/jades/1001 Intro to Data Science Notebooks/Project/wildfires-1001/'

subdir_from = 'data/clean_data/mapping_tables/'
filename_from = 'station_grid_map.csv'

station_grid_map = pd.read_csv(gitdir + subdir_from + filename_from)
```

```python
df_noaa = station_grid_map.merge(df_master_trim, how='left', left_on='ID', right_on='ID')
```

```python
#Save the cleaned dataframe
subdir_to = 'data/clean_data/NOAA/'
filename_to = 'noaa_grid_data.csv'
df_noaa.to_csv(gitdir + subdir_to + filename_to, index=False)
```
